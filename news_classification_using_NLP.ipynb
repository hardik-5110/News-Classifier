```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
import nltk
nltk.download('all')
```

```python
fake = pd.read_csv('/content/Fake.csv')
```

```python
genuine = pd.read_csv('/content/True.csv')
```

```python
fake=fake.iloc[:,0:4]
fake
```

```python
genuine
```

```python
fake["target"]=0
genuine["target"]=1
```

```python
fake.head()
```

```python
genuine.head()
```

```python
data = pd.concat([fake,genuine],axis=0)
```

```python
data
```

```python
data = data.reset_index(drop=True)   # numeric indexing 
data
```

```python
data = data.drop(['title','subject','date'],axis=1)
```

```python
data
```

```python
# Data Preprocessing
```

```python
#tokenization
from nltk.tokenize import word_tokenize
data['text'] = data['text'].apply(word_tokenize)
```

```python
data.head()
```

```python
#stemming
from nltk.stem.snowball import SnowballStemmer
porter = SnowballStemmer("english")
```

```python
def stem_it(text):
    return(porter.stem(word) for word in text)
```

```python
data['text'] = data['text'].apply(stem_it)
```

```python
data.head()
```

```python
# stop word removal
def stop_it(t):
  dt=[word for word in t if len(word)>2]
  return dt
```

```python
data['text'] = data['text'].apply(stop_it)
```

```python
data.head(10)
```

```python
data['text'] = data['text'].apply(' '.join)
```

```python
# Splitting data Set
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data['text'],data['target'],test_size=0.25) 
```

```python
display(X_train.head())
print('\n')
display(y_train.head())
```

```python
# Vectorization (TFIDF) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_df=0.7)

tfidf_train = tfidf.fit_transform(X_train)
tfidf_test = tfidf.transform(X_test) 
```

```python
print(tfidf_train)
```

```python
# Building of ML model
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
model1= LogisticRegression(max_iter=900)
model1.fit(tfidf_train,y_train)
```

```python
pred1 = model1.predict(tfidf_test)
```

```python
pred1
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
cr1 = accuracy_score(y_test,pred1)
cr1
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier

model2 =  PassiveAggressiveClassifier(max_iter=100)
model2.fit(tfidf_train,y_train)
```

```python
pred2 = model2.predict(tfidf_test)
```

```python
cr2 = accuracy_score(y_test,pred2)
cr2
```
